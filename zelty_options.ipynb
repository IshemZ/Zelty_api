```python
import pandas as pd
import json
from zelty_orders import options
import ast
from pprint import pprint
```

```python
option_list = options
option_list
pprint(option_list)
```

```python
# Extraire les plats
option_list = option_list['options']
option_list
```

```python
option_list = pd.DataFrame(option_list)
option_list.head()
```

```python
def clean_data(option_list):
    # Supprimer les colonnes : 'remote_id', 'disable' et 6 autres colonnes
    option_list = option_list.drop(columns=['remote_id', 'disable', 'zc_only', 'donotpresent', 'meta', 'pick_same_value', 'min_choices', 'max_choices'])
    return option_list

option_list = clean_data(option_list.copy())
option_list.head()
```

```python
option_list.to_csv('../Zelty_api/Data/option_list.csv', index=False)
```

```python
df_exploded = option_list.explode('values')
df_exploded.head()
```

```python
# Convertir les dictionnaires de la colonne 'parts' en colonnes
df_values = pd.json_normalize(df_exploded['values'])
df_values.head()
```

```python
columns_to_drop = [
        'remote_id', 'description', 'image', 'price'
    ]
```

```python
def clean_values(df_values):
    # Supprimer toutes les colonnes inutiles en une seule opération
    df_values = df_values.drop(columns=columns_to_drop)
    return df_values
    
df_values_clean = clean_values(df_values.copy())
df_values_clean.head()
```

```python
df_values_clean.to_csv('../Zelty_api/Data/zelty_option.csv', index=False)
```

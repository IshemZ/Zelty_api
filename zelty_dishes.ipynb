```python
columns_to_drop = [
    'remote_id', 'id_restaurant', 'sku', 'image', 'thumb', 
    'price_togo', 'happy_price', 'cost_price', 'price_delivery', 
    'tvad', 'tvat', 'tax_takeaway', 'tax_delivery', 
    'loyalty_points', 'color', 'earn_loyalty', 'loyalty_points_discount', 
    'fab_name', 'price_to_define', 'disable_before', 'disable_after', 
    'zc_name', 'meta', 'zc_only', 'tax', 'tva', 'o',
    'weight_for_price', 'disable', 'id_fabrication_place', 'disable_takeaway','disable_delivery'
]
```

```python
def clean_data(df):
    # Supprimer toutes les colonnes inutiles en une seule op√©ration
    df = df.drop(columns=columns_to_drop)
    df = df.astype({'id': 'object'},{'price': 'float64'})
    df['price'] = df['price'] / 100
    df['tags'] = df['tags'].apply(lambda x: ', '.join(map(str, x)) if isinstance(x, list) else x)
    df['options'] = df['options'].apply(lambda x: ', '.join(map(str, x)) if isinstance(x, list) else x)
    return df

df_clean = clean_data(df.copy())
df_clean.head()
```
